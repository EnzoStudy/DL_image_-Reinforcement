```python
img1 = cv.imread("/content/68562d1313642314 hd wallpapers 1080p hd wallpapers 1080p img 1080p.jpg")
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(img1)

cv.waitKey(0)
```

```python
height = img1.shape[0]
width = img1.shape[1]

print(img1.shape)
```

```python
import numpy as np
size_int = 4

resize_img = np.zeros((height,width,3),dtype=np.uint8)
small_img = np.zeros((int(height/size_int),int(width/size_int),3),dtype=np.uint8)

#cv2_imshow(resize_img)
#cv.waitKey(0)
```

```python
for i in range(height):
  for j in range(width):
    if i % size_int==0 or j %size_int ==0:
      small_img[int(i/size_int),int(j/size_int)] = img1[i,j]


```

```python
cv2_imshow(small_img)
cv.waitKey(0)
```

```python
print(small_img.shape)

cv.imwrite("/content/small.jpg", small_img)

```
